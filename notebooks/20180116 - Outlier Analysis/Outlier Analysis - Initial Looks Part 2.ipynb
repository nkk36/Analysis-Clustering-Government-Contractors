# Company Looks

Now we look back at the same 5 companies we looked at last [time](https://github.com/nkk36/Analysis-Clustering-Government-Contractors/blob/master/notebooks/20180116%20-%20Outlier%20Analysis/Outlier%20Analysis%20-%20Initial%20Looks.ipynb). They were Lockheed Martin, Cardinal Health 200, Owens and Minor, McKeeson Corporation, and SAIC. 

Instead of counting the number of unique PIIDs per NAICS code this time we just add the dollars obligated per NAICS code for each company. We also calculate the percentage of total revenue each NAICS code constitutes. 

## Lockheed Martin

Last time we saw that most of Lockheed's unique PIIDs were for machine shops which are used to support their primary enterprise - aircraft manufacturing. We sort the list in decreasing order by total dollars obligated per NAICS code.

This time when we add dollars obligated we see that a significant portion of their revenue (38%) come from aircraft manufacturing. The second largest activity in terms of dollars obligated is the manufacturing of missiles and space vehicles at 13.5%. We don't actually see machine shops until way down the list at row 49. **Machine shops only make up 0.06% of their revenue whereas before they represented 95% of the unique PIIDs**. Already this seems to be giving a better representation of what Lockheed Martin does.

```python
Lockheed = df_original.loc[df_original.parentdunsnumber == 834951691,].groupby(["principalnaicscode"]).dollarsobligated.sum().reset_index()
Lockheed.sort_values(by = "dollarsobligated",
                     ascending = False,
                     inplace = True)
Lockheed = Lockheed.reset_index(drop = True)
Lockheed["Percentage"] = 100*Lockheed["dollarsobligated"]/sum(Lockheed["dollarsobligated"])
Lockheed.head(10)
```

```python
Lockheed.iloc[40:50,:]
```

## Cardinal Health

Almost 55% of Cardinal Health's revenue comes from 423450: Medical, Dental, and Hospital Equipment and Supplies Merchant Wholesalers. This was at the top of their list last time, but this time 339112: Surgical and medical instrument manufacturing comes second making up 41% of their revenue. In terms of the NAICS codes we are seeing, this seems to be consistent with what we saw last time.

```python
Cardinal_Health = df_original.loc[df_original.parentdunsnumber == 97537435,].groupby(["principalnaicscode"]).dollarsobligated.sum().reset_index()
Cardinal_Health.sort_values(by = "dollarsobligated",
                     ascending = False,
                     inplace = True)
Cardinal_Health = Cardinal_Health.reset_index(drop = True)
Cardinal_Health["Percentage"] = 100*Cardinal_Health["dollarsobligated"]/sum(Cardinal_Health["dollarsobligated"])
Cardinal_Health.head(10)
```
